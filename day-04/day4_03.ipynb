```python

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import KNNImputer,SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/iamdipsan/100-Days-ML/refs/heads/main/day-39/train.csv", usecols=['Age', 'Pclass', 'Fare', 'Survived'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
X=df.drop(columns=['Survived'])
y=df['Survived']
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train.head()
```

```python
knn=KNNImputer(n_neighbors=3,weights='distance')
X_train_transformed=knn.fit_transform(X_train)
X_test_transformed=knn.transform(X_test)
```

```python
lr=LogisticRegression()
lr.fit(X_train_transformed,y_train)
```

```python
y_pred=lr.predict(X_test_transformed)
accuracy_score(y_test,y_pred)
```

```python
# Comparision with Simple Imputer --> mean

si = SimpleImputer()

X_train_trf2 = si.fit_transform(X_train)
X_test_trf2 = si.transform(X_test)
```

```python
lr = LogisticRegression()

lr.fit(X_train_trf2,y_train)

y_pred2 = lr.predict(X_test_trf2)

accuracy_score(y_test,y_pred2)
```
