```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/iamdipsan/100-Days-ML/refs/heads/main/day-33/titanic.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['number'].unique()
```

```python
fig=df['number'].value_counts().plot.bar()
fig.set_title('Passenger travelling with')
```

```python
df['number_numerical']=pd.to_numeric(df['number'], errors='coerce',downcast='integer')
```

```python
df['number_categorical']=np.where(df['number_numerical'].isnull(), df['number'],np.nan)
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python

# Capture the numerical part
df['cabin_num'] = df['Cabin'].str.extract(r'(\d+)')

# Capture the first letter
df['cabin_cat'] = df['Cabin'].str.extract(r'([A-Za-z])')
```

```python
df.head()
```

```python
df['cabin_cat'].value_counts()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
df['Ticket']
```

```python

# Ensure 'Ticket' column has no NaN before processing
df['ticket_num'] = df['Ticket'].str.split().str[-1]  # Extract last part
df['ticket_num'] = pd.to_numeric(df['ticket_num'], errors='coerce', downcast='integer')

# Extract first part of ticket as category
df['ticket_cat'] = df['Ticket'].str.split().str[0]

# If 'ticket_cat' is fully numeric, replace it with NaN
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan, df['ticket_cat'])
```
