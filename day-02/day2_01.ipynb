```python
df=df.iloc[:,2:]
```

```python
X=df.drop(columns='purchased')
y=df['purchased']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=  train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from  sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
oe.fit(X_train)
X_train_transformed=oe.transform(X_train)
X_test_transformed=oe.transform(X_test)
```

```python
X_train_transformed = pd.DataFrame(X_train_transformed, columns=['review', 'education'])
X_test_transformed = pd.DataFrame(X_test_transformed, columns=['review', 'education'])
```
