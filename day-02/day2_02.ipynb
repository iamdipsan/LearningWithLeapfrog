```python

import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/iamdipsan/100-Days-ML/refs/heads/main/day-27/cars.csv")
```

```python
df.head()
```

```python
df['brand'].value_counts()
```

```python
X=df.drop(columns='selling_price')
y=df['selling_price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop='first', sparse_output=False)  # use sparse_output instead of sparse

ohe.fit(X_train[['fuel']])
X_train_transformed = ohe.transform(X_train[['fuel']])
X_test_transformed = ohe.transform(X_test[['fuel']])
```

```python
column_names = ohe.get_feature_names_out(['fuel'])
```

```python
column_names
```

```python
X_train_transformed=pd.DataFrame(X_train_transformed,columns=column_names)
X_test_transformed=pd.DataFrame(X_test_transformed,columns=column_names)
```

```python
X_train_transformed.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
model = RandomForestRegressor(random_state=42)

# Fit model on training data
model.fit(X_train_transformed, y_train)

# Predict on test data
y_pred = model.predict(X_test_transformed)

# Evaluate
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```

```python

X_predict=['Maruti',1000,1.0,'First Owner']
result=model.predict(X_predict)
print(result)
```
